```python
import matplotlib.pyplot as plt

# Donn√©es
hours = ['14h', '15h', '16h', '17h', '18h']
wifi_bouygues = [10, 11, 11, 11, 11]
ethernet_bouygues = [8, 8, 8, 8, 8]
fiveg = [39, 37, 43, 44, 41]
fourg = [42, 48, 50, 44, 51]

# Trac√© des donn√©es
plt.figure(figsize=(10, 6))
plt.plot(hours, wifi_bouygues, marker='o', label='Wi-Fi Bouygues')
plt.plot(hours, ethernet_bouygues, marker='o', label='Ethernet Bouygues')
plt.plot(hours, fiveg, marker='o', label='5G')
plt.plot(hours, fourg, marker='o', label='4G')

# Ajouter des labels et un titre
plt.xlabel('Hour')
plt.ylabel('Latency (ms)')
plt.title('Network Latency by Connection Type')
plt.legend()
plt.grid(True)

# Affichage du graphique
plt.show()
```
